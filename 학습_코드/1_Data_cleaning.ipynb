## 2-2. 정제(Cleaning)
- 정제(cleaning) : 갖고 있는 코퍼스(말뭉치)로부터 노이즈 데이터를 제거한다.

### 전처리 사용자 정의 함수(최종 & 불용어 리스트 2차 필터)

```python
# 텍스트 전처리 사용자 정의함수(UDF of text pre-processing)
def text_preprocessor(s):
    import re

    ## (1) 이메일, 링크
    pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'  # http
    s = re.sub(pattern=pattern, repl=' ', string=s)

    pattern = r'(?<=\S)www(?=\s|$)|(?<=\s)www(?=\s|$)'  # www
    s = re.sub(pattern=pattern, repl=' ', string=s)

    pattern = r'(?<=\S)@(?=\s|$)|(?<=\s)@(?=\s|$)'  # @
    s = re.sub(pattern=pattern, repl=' ', string=s)

    ## (2) 특수문자, 숫자
    pattern = r'[^a-zA-Z가-힣]'
    s = re.sub(pattern=pattern, repl=' ', string=s)

    ## (3) 단위, 기타(‘\n’,’\r’, ‘#’)
    units = ['mm', 'cm', 'km', 'ml', 'kg', 'g']
    for unit in units:
        s = s.lower()  # 대문자를 소문자로 변환
        s = re.sub(pattern=unit, repl='', string=s)

    pattern = r'[\n\r#]'
    s = re.sub(pattern=pattern, repl=' ', string=s)

    ## (4) 하나의 자음, 모음(ex.ㅋㅋㅋ), 한 글자
    pattern = r'(?![ㄱ-ㅣ가-힣])\w{1}'
    s = re.sub(pattern=pattern, repl='', string=s)

    # (5) 작성자
    # 작성자 컬럼의 고유한 값 추출
    unique_authors = youth_sns_df['작성자'].unique()

    # 작성자 컬럼과 제목, 내용 컬럼을 비교하여 동일한 내용 제외
    s = re.sub(rf'(?i)\b(?:{"|".join(map(re.escape, unique_authors))})\b', '', s)

    return s

# 완전 중복 행 제거
youth_sns_df = youth_sns_df.drop_duplicates()
youth_sns_df = youth_sns_df.reset_index(drop=True)

# '내용' 컬럼에 대해 전처리 함수 적용
youth_sns_df['clean_text'] = youth_sns_df['내용'].apply(text_preprocessor)

# 결과 출력
youth_sns_df
```
