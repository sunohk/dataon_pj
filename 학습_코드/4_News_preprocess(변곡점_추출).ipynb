```python
#create a function to subset dates

#날짜별 peak 분석
# nurse  = 2022/1/09-11, 2023/05/15-17
# stalking = 2020/04/1-3, 2022/09/15-17
# rent = 2023/01/10-12, 2023/04/18-20
# youth = 2021/04/07-09, 2022/03/09-11

#subset based on dates
def subset_df(df, start_date, end_date):
  subset_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
  subset_df = subset_df.sort_values(by='Comments', ascending=False).head(3)
  return subset_df

# #choose top 3 articles with the most comments
```

```python
# subset def

# Create a dictionary with DataFrames and date ranges
dataframes = {
'youth_peak1':youth_peak1,
'youth_peak2':youth_peak2,
'rent_peak1':rent_peak1,
'rent_peak2' :rent_peak2,
'stalking_peak1':stalking_peak1,
'stalking_peak2':stalking_peak2,
'nurse_peak1':nurse_peak1,
'nurse_peak2':nurse_peak2,
}

date_ranges = {
'youth_peak1' : ('2021-04-07', '2021-04-09'),
'youth_peak2': ('2022-03-09', '2022-03-11'),
'rent_peak1':  ('2023-01-10', '2023-01-12'),
'rent_peak2' : ('2023-04-18', '2023-04-20'),
'stalking_peak1': ('2020-04-01', '2020-04-03'),
'stalking_peak2': ('2022-09-15', '2022-09-17'),
'nurse_peak1': ('2022-01-09', '2022-01-11'),
'nurse_peak2': ('2023-05-15', '2023-05-17'),
}

# Apply the subset_df function to different DataFrames with different date ranges
results = pd.DataFrame()

for name, df in dataframes.items():
    start_date, end_date = date_ranges[name]
    result = subset_df(df, start_date, end_date)
    result['Topic'] = name
    results = pd.concat([results, result])

results
```
