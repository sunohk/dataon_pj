## 2. 여론 시각화(월 기준)

```python
# 'Date' 컬럼을 날짜로 변환
youth_df['Date'] = pd.to_datetime(youth_df['Date'], errors='coerce')

# 'Date'를 기준으로 연도와 월을 추출
youth_df['Year'] = youth_df['Date'].dt.year
youth_df['Month'] = youth_df['Date'].dt.month

# label이 0인 데이터와 1인 데이터로 분리
label_0_df = youth_df[youth_df['감성분석'] == 0]
label_1_df = youth_df[youth_df['감성분석'] == 1]

# 연도별 월별 label 0과 label 1의 개수 계산
label_0_counts = label_0_df.groupby(['Year', 'Month']).size().reset_index(name='Negative_Count')
label_1_counts = label_1_df.groupby(['Year', 'Month']).size().reset_index(name='Positive_Count')

# 연도와 월을 결합하여 'Year-Month' 컬럼 생성
label_0_counts['Year-Month'] = label_0_counts['Year'].astype(str) + '-' + label_0_counts['Month'].astype(str)
label_1_counts['Year-Month'] = label_1_counts['Year'].astype(str) + '-' + label_1_counts['Month'].astype(str)
# Create a Plotly figure and add traces
fig = go.Figure()

fig.add_trace(go.Scatter(x=label_0_counts['Year-Month'], y=label_0_counts['Negative_Count'], mode='lines+markers', name='Negative', line=dict(color='red')))
fig.add_trace(go.Scatter(x=label_1_counts['Year-Month'], y=label_1_counts['Positive_Count'], mode='lines+markers', name='Positive', line=dict(color='blue')))

# Update layout if needed
fig.update_layout(title='Monthly Changes in Label Count')
fig.update_xaxes(tickangle=45)

# Show the Plotly figure
fig.show()
```

## 3. 변곡점 추출

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from scipy.signal import find_peaks

# 'Year-Month' 컬럼 생성
youth_df['Year-Month'] = youth_df['Year'].astype(str) + '-' + youth_df['Month'].astype(str).str.zfill(2)

# 월별로 데이터 개수를 카운트
monthly_counts = youth_df.groupby('Year-Month').size().reset_index(name='월별데이터수')

# 월별 데이터 개수의 평균 계산
threshold = monthly_counts['월별데이터수'].mean()

# 양의 변곡점 찾기
peaks, _ = find_peaks(monthly_counts['월별데이터수'], height=threshold)

# 찾은 변곡점을 시각화
fig = go.Figure()
fig.add_trace(go.Scatter(x=monthly_counts['Year-Month'], y=monthly_counts['월별데이터수'], mode='lines', name='월별데이터수'))
fig.add_trace(go.Scatter(x=monthly_counts['Year-Month'][peaks], y=monthly_counts['월별데이터수'][peaks], mode='markers', name='월별데이터수 Peaks', marker=dict(color='red', size=8)))

# 임계값 표시
fig.add_shape(go.layout.Shape(type='line', x0=monthly_counts['Year-Month'].iloc[0], x1=monthly_counts['Year-Month'].iloc[-1], y0=threshold, y1=threshold, line=dict(dash='dash', color='gray')))

# 그래프 레이아웃 설정
fig.update_layout(title='월별 데이터 개수 변곡점 탐지', xaxis_title='Year-Month', yaxis_title='월별데이터수')

# 그래프 표시
fig.show()
```

## 4 긍부정 시각화 타입별로

```python
import pandas as pd
import plotly.graph_objects as go

# 각 타입별로 데이터 분리
types = ['트위터', '인스타그램', '커뮤니티', '블로그']

# Create traces for each type and sentiment (Twitter, Instagram, Community, Blog)
fig = go.Figure()

colors = ['red', 'blue', 'green', 'purple', 'orange', 'pink', 'cyan', 'brown']

for idx, type_name in enumerate(types):
    for sentiment in [0, 1]:
        # 각 타입별로 데이터 추출
        type_df = label_0_df[label_0_df['Type'] == type_name] if sentiment == 0 else label_1_df[label_1_df['Type'] == type_name]

        # 연도별 월별 빈도수 계산
        counts = type_df.groupby(['Year', 'Month']).size().reset_index(name=f'{type_name} ({ "Negative" if sentiment == 0 else "Positive" })')

        # Trace 추가
        fig.add_trace(go.Scatter(
            x=counts['Year'].astype(str) + '-' + counts['Month'].astype(str),
            y=counts[f'{type_name} ({ "Negative" if sentiment == 0 else "Positive" })'],
            mode='lines+markers',
            name=f'{type_name} ({ "Negative" if sentiment == 0 else "Positive" })',
            line=dict(color=colors[idx * 2 + sentiment])
        ))

# Create the layout for the plot
layout = go.Layout(
    title='Monthly Changes in Sentiment Count by Type',
    xaxis=dict(title='Year-Month', tickangle=45),
    yaxis=dict(title='Sentiment Count'),
    legend=dict(x=0, y=1)
)

# Update layout and show the interactive plot
fig.update_layout(layout)
fig.show()
```

## 6 언급량 최다 기록 분석

```python
import pandas as pd
import plotly.express as px

def plot_daily_label_counts_by_month(df, year, month):
    # Filter the DataFrame for the specified year and month
    filtered_data = df[(df['Year'] == year) & (df['Month'] == month)]

    # Calculate daily counts
    daily_counts = filtered_data.groupby('Date')['감성분석'].value_counts().unstack().fillna(0)

    # Create a DataFrame for daily label counts
    daily_counts_df = pd.DataFrame({
        'Date': daily_counts.index,
        'Positive': daily_counts.get(1, 0),
        'Negative': daily_counts.get(0, 0)
    })

    # Create a line plot using Plotly
    fig = px.line(daily_counts_df, x='Date', y=['Positive', 'Negative'], markers=True,
                  title=f'Daily Changes in Label Count for {year}-{month:02d}')
    fig.update_xaxes(title='Date', tickangle=45)
    fig.update_yaxes(title='Label Count')
    fig.show()

```

```python
# 'Year'와 'Month'를 기준으로 'Negative_Count' 합계 구하기
negative_counts_sum = label_0_counts.groupby(['Year', 'Month'])['Negative_Count'].sum().reset_index()

# 'Year'와 'Month'를 기준으로 'Positive_Count' 합계 구하기
positive_counts_sum = label_1_counts.groupby(['Year', 'Month'])['Positive_Count'].sum().reset_index()

# 'Negative_Count'와 'Positive_Count'를 합한 값을 새로운 열로 추가
negative_counts_sum['Total_Count'] = negative_counts_sum['Negative_Count'] + positive_counts_sum['Positive_Count']

# 'Total_Count'에서 최대값을 가지는 행 찾기
max_total_count_row = negative_counts_sum.loc[negative_counts_sum['Total_Count'].idxmax()]

# 최대값을 가지는 연도와 월 찾기
max_year = max_total_count_row['Year']
max_month = max_total_count_row['Month']

# 최대값 출력
print(f"최대 합계를 가지는 연도: {max_year}, 월: {max_month}")
```

```python
import pandas as pd

def filter_dataframe_by_date_range(dataframe, start_date, end_date, date_column='작성일', date_format='%Y-%m-%d'):
    """
    Filter a DataFrame based on a date range.

    Parameters:
        dataframe (pd.DataFrame): The DataFrame to filter.
        start_date (str): The start date of the range in 'yyyy-mm-dd' format.
        end_date (str): The end date of the range in 'yyyy-mm-dd' format.
        date_column (str): The name of the column containing dates.
        date_format (str): The date format of the date_column.

    Returns:
        pd.DataFrame: A filtered DataFrame containing rows within the specified date range.
    """
    # Convert date strings to datetime objects
    dataframe[date_column] = pd.to_datetime(dataframe[date_column], format=date_format, errors='coerce')

    # Filter the DataFrame based on the date range
    filtered_data = dataframe[(dataframe[date_column] >= start_date) & (dataframe[date_column] <= end_date)]

    return filtered_data
```
